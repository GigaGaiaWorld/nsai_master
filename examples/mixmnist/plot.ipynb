```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.patches import Rectangle
import seaborn as sns

def create_loss_timeline_chart(file_paths=None, save_path=None):
    """
    创建多语言模型训练Loss时间序列图表
    
    Args:
        file_paths (dict): 包含文件路径的字典，格式为 {'arabia': 'path1', 'iraq': 'path2', 'pakistan': 'path3'}
        save_path (str): 保存图片的路径，如果为None则直接显示
    
    Returns:
        matplotlib.figure.Figure: 图表对象
    """
    
    # 如果没有提供文件路径，使用示例数据
    if file_paths is None:
        # 示例数据（从您提供的文件中提取）
        arabia_data = {
            'time': [3.99, 6.49, 9.13, 11.65, 14.18, 16.72],
            'loss': [0.3878, 0.2178, 0.2070, 0.2091, 0.2352, 0.1401],
            'accuracy': 0.8867
        }
        
        iraq_data = {
            'time': [4.16, 6.27, 8.30, 10.32, 12.35, 14.37],
            'loss': [0.4552, 0.3563, 0.2240, 0.2089, 0.2872, 0.1398],
            'accuracy': 0.94
        }
        
        pakistan_data = {
            'time': [4.05, 7.95, 11.42, 14.87, 18.31, 21.73, 25.26, 28.76, 32.18, 35.58, 39.12, 42.66, 46.07, 49.68, 53.12, 56.54],
            'loss': [0.0559, 0.0100, 0.0100, 0.0100, 0.0100, 0.0560, 0.0100, 0.0100, 0.0100, 0.1479, 0.1020, 0.0560, 0.0100, 0.0100, 0.0100, 0.0100],
            'accuracy': 0.9915
        }
    else:
        # 从文件读取数据的函数
        def parse_log_file(file_path):
            with open(file_path, 'r') as f:
                content = f.read()
            
            # 提取准确率
            accuracy_line = [line for line in content.split('\n') if line.startswith('#Accuracy')][0]
            accuracy = float(accuracy_line.split(' ')[1])
            
            # 提取CSV数据
            csv_start = content.find('i,time,loss')
            csv_data = content[csv_start:]
            
            # 解析CSV
            lines = csv_data.strip().split('\n')[1:]  # 跳过header
            times, losses = [], []
            for line in lines:
                parts = line.split(',')
                times.append(float(parts[1]))
                losses.append(float(parts[2]))
            
            return {'time': times, 'loss': losses, 'accuracy': accuracy}
        
        arabia_data = parse_log_file(file_paths['arabia'])
        iraq_data = parse_log_file(file_paths['iraq'])
        pakistan_data = parse_log_file(file_paths['pakistan'])
    
    # 设置图表样式
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(1, 1, figsize=(16, 10))
    
    # 定义颜色
    colors = {
        'arabia': '#ff6b6b',
        'iraq': '#4ecdc4', 
        'pakistan': '#45b7d1'
    }
    
    # 计算时间偏移
    arabia_end_time = max(arabia_data['time'])
    iraq_end_time = max(iraq_data['time'])
    total_time = arabia_end_time + iraq_end_time + max(pakistan_data['time'])
    
    # 调整时间轴使其连续
    iraq_times_adjusted = [t + arabia_end_time for t in iraq_data['time']]
    pakistan_times_adjusted = [t + arabia_end_time + iraq_end_time for t in pakistan_data['time']]
    
    # 获取初始和最终loss值
    arabia_initial_loss = arabia_data['loss'][0]
    arabia_final_loss = arabia_data['loss'][-1]
    iraq_initial_loss = iraq_data['loss'][0]
    iraq_final_loss = iraq_data['loss'][-1]
    pakistan_initial_loss = pakistan_data['loss'][0]
    
    # 绘制背景区域来区分不同阶段
    ax.axvspan(0, arabia_end_time, alpha=0.1, color=colors['arabia'], label='Arabia-Train')
    ax.axvspan(arabia_end_time, arabia_end_time + iraq_end_time, alpha=0.1, color=colors['iraq'], label='Iraq-Train')
    ax.axvspan(arabia_end_time + iraq_end_time, total_time, alpha=0.1, color=colors['pakistan'], label='Pakistan-Train')
    
    # 绘制背景基线（虚线）
    # 阶段1: Iraq和Pakistan的初始值
    ax.plot([0, arabia_end_time], [iraq_initial_loss, iraq_initial_loss], 
            '--', color=colors['iraq'], alpha=0.4, linewidth=2, label='_nolegend_')
    ax.plot([0, arabia_end_time], [pakistan_initial_loss, pakistan_initial_loss], 
            '--', color=colors['pakistan'], alpha=0.4, linewidth=2, label='_nolegend_')
    
    # 阶段2: Arabia最终值和Pakistan初始值
    ax.plot([arabia_end_time, arabia_end_time + iraq_end_time], [arabia_final_loss, arabia_final_loss], 
            '--', color=colors['arabia'], alpha=0.4, linewidth=2, label='_nolegend_')
    ax.plot([arabia_end_time, arabia_end_time + iraq_end_time], [pakistan_initial_loss, pakistan_initial_loss], 
            '--', color=colors['pakistan'], alpha=0.4, linewidth=2, label='_nolegend_')
    
    # 阶段3: Arabia和Iraq的最终值
    ax.plot([arabia_end_time + iraq_end_time, total_time], [arabia_final_loss, arabia_final_loss], 
            '--', color=colors['arabia'], alpha=0.4, linewidth=2, label='_nolegend_')
    ax.plot([arabia_end_time + iraq_end_time, total_time], [iraq_final_loss, iraq_final_loss], 
            '--', color=colors['iraq'], alpha=0.4, linewidth=2, label='_nolegend_')
    
    # 绘制主要训练曲线
    ax.plot(arabia_data['time'], arabia_data['loss'], 
            color=colors['arabia'], linewidth=3, marker='o', markersize=6,
            label=f'Arabia (Accuracy: {arabia_data["accuracy"]:.2%})')
    
    ax.plot(iraq_times_adjusted, iraq_data['loss'], 
            color=colors['iraq'], linewidth=3, marker='s', markersize=6,
            label=f'Iraq (Accuracy: {iraq_data["accuracy"]:.2%})')
    
    ax.plot(pakistan_times_adjusted, pakistan_data['loss'], 
            color=colors['pakistan'], linewidth=3, marker='^', markersize=5,
            label=f'Pakistan (Accuracy: {pakistan_data["accuracy"]:.2%})')
    
    # 添加阶段分隔线
    ax.axvline(x=arabia_end_time, color='gray', linestyle='-', alpha=0.6, linewidth=1)
    ax.axvline(x=arabia_end_time + iraq_end_time, color='gray', linestyle='-', alpha=0.6, linewidth=1)
    
    # 添加阶段标注
    ax.text(arabia_end_time/2, ax.get_ylim()[1]*0.9, 'Stage1: Arabia-Train', 
            ha='center', va='center', fontsize=12, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor=colors['arabia'], alpha=0.2))
    
    ax.text(arabia_end_time + iraq_end_time/2, ax.get_ylim()[1]*0.9, 'Stage2: Iraq-Train', 
            ha='center', va='center', fontsize=12, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor=colors['iraq'], alpha=0.2))
    
    ax.text(arabia_end_time + iraq_end_time + (total_time - arabia_end_time - iraq_end_time)/2, 
            ax.get_ylim()[1]*0.9, 'Stage3: Pakistan-Train', 
            ha='center', va='center', fontsize=12, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor=colors['pakistan'], alpha=0.2))
    
    # 设置图表属性
    ax.set_xlabel('Cumulative time (seconds)', fontsize=14, fontweight='bold')
    ax.set_ylabel('Loss', fontsize=14, fontweight='bold')
    ax.set_title('Graph of multilingual model training over time', fontsize=18, fontweight='bold', pad=20)
    
    # 设置图例
    ax.legend(loc='upper right', frameon=True, fancybox=True, shadow=True, fontsize=12)
    
    # 设置网格
    ax.grid(True, alpha=0.3)
    ax.set_axisbelow(True)
    
    # 设置y轴从0开始
    ax.set_ylim(bottom=0)
    
    # 添加最终loss值的标注
    ax.annotate(f'Final Loss: {arabia_final_loss:.3f}', 
                xy=(arabia_data['time'][-1], arabia_final_loss),
                xytext=(10, 10), textcoords='offset points',
                bbox=dict(boxstyle="round,pad=0.3", facecolor=colors['arabia'], alpha=0.3),
                arrowprops=dict(arrowstyle='->', color=colors['arabia']),
                fontsize=10)
    
    ax.annotate(f'Final Loss: {iraq_final_loss:.3f}', 
                xy=(iraq_times_adjusted[-1], iraq_final_loss),
                xytext=(10, 10), textcoords='offset points',
                bbox=dict(boxstyle="round,pad=0.3", facecolor=colors['iraq'], alpha=0.3),
                arrowprops=dict(arrowstyle='->', color=colors['iraq']),
                fontsize=10)
    
    ax.annotate(f'Final Loss: {pakistan_data["loss"][-1]:.3f}', 
                xy=(pakistan_times_adjusted[-1], pakistan_data['loss'][-1]),
                xytext=(10, 10), textcoords='offset points',
                bbox=dict(boxstyle="round,pad=0.3", facecolor=colors['pakistan'], alpha=0.3),
                arrowprops=dict(arrowstyle='->', color=colors['pakistan']),
                fontsize=10)
    
    # 调整布局
    plt.tight_layout()
    
    # 保存或显示图表
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"图表已保存到: {save_path}")
    else:
        plt.show()
    
    return fig

# 使用示例
if __name__ == "__main__":
    # 使用示例数据
    fig = create_loss_timeline_chart()

    # 或者使用文件路径
    # file_paths = {
    #     'arabia': 'language_arabia_2.log',
    #     'iraq': 'language_iraq_2.log', 
    #     'pakistan': 'language_pakistan_2.log'
    # }
    # fig = create_loss_timeline_chart(file_paths, save_path='loss_timeline.png')
```
