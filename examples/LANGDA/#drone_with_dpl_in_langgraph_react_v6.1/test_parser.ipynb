```python
from problog.program import PrologString
from problog.logic import Term, Clause, Constant, Var, is_variable

def parse_term(term, depth=0):
    """递归解析任何 Term 对象及其子项"""
    indent = "  " * depth
    
    # 打印基本信息
    print(f"{indent}Term: {term}")
    print(f"{indent}类型: {type(term).__name__}")
    
    # 检查类型，不同的类型有不同的属性
    if isinstance(term, Term):
        print(f"{indent}函子: {term.functor}")
        print(f"{indent}参数数量: {term.arity}")
        
        # 检查是否有概率标注
        if hasattr(term, 'probability') and term.probability is not None:
            print(f"{indent}概率: {term.probability}")
        
        # 递归解析所有参数
        if term.arity > 0:
            print(f"{indent}参数列表:")
            for i, arg in enumerate(term.args):
                print(f"{indent}  参数 {i+1}:")
                parse_term(arg, depth+2)
    
    # 如果是变量，打印变量信息
    elif is_variable(term):
        print(f"{indent}变量名: {term.name}")
        # 变量可能有其他属性，但不一定有id
        if hasattr(term, 'id'):
            print(f"{indent}变量ID: {term.id}")
    
    # 如果是常量，打印常量信息
    elif isinstance(term, Constant):
        print(f"{indent}常量值: {term.value}")
        if hasattr(term, 'typ'):
            print(f"{indent}常量类型: {term.typ}")
    
    # 其他类型
    else:
        print(f"{indent}值: {term}")
    
    print(f"{indent}-----")

def parse_body(body, depth=0):
    """特殊处理体部，因为它可能包含复杂的逻辑连接词"""
    indent = "  " * depth
    
    if body is None:
        print(f"{indent}体部为空")
        return
    
    print(f"{indent}体部: {body}")
    
    # 检查是否是Term类型
    if isinstance(body, Term):
        # 检查是否是连词 (AND)
        if body.functor == 'conjunction':
            print(f"{indent}这是一个AND连词，包含 {len(body.args)} 个子目标")
            for i, subgoal in enumerate(body.args):
                print(f"{indent}  子目标 {i+1}:")
                parse_term(subgoal, depth+2)
        
        # 检查是否是析取 (OR)
        elif body.functor == 'disjunction':
            print(f"{indent}这是一个OR连词，包含 {len(body.args)} 个选择")
            for i, choice in enumerate(body.args):
                print(f"{indent}  选择 {i+1}:")
                parse_term(choice, depth+2)
        
        # 其他情况（单个目标）
        else:
            parse_term(body, depth+1)
    else:
        print(f"{indent}体部不是Term类型，而是: {type(body).__name__}")
        parse_term(body, depth+1)

def parse_clause(clause, depth=0):
    """解析一个完整的子句（包括头部和体部）"""
    indent = "  " * depth
    
    print(f"{indent}子句: {clause}")
    print(f"{indent}类型: {type(clause).__name__}")
    
    # 确保子句有头部
    if hasattr(clause, 'head'):
        print(f"{indent}头部:")
        parse_term(clause.head, depth+1)
    else:
        print(f"{indent}警告: 子句没有头部属性")
    
    # 解析体部
    if hasattr(clause, 'body'):
        print(f"{indent}体部:")
        parse_body(clause.body, depth+1)
    else:
        print(f"{indent}警告: 子句没有体部属性")

# 创建一个 ProbLog 程序，包含各种结构
prog = PrologString("""
    0.7::heads.
    coin(c1).
    coin(c2).
    toss(X) :- coin(X).
    parent(john, mary).
    grandparent(X, Z) :- parent(X, Y), parent(Y, Z).
    sibling(X, Y) :- parent(Z, X), parent(Z, Y), X \== Y.
    path(X, Y) :- edge(X, Y).
    path(X, Z) :- edge(X, Y), path(Y, Z).
    happy(X) :- rich(X); famous(X).
""")

# 遍历并解析每个语句
print("开始全面解析 ProbLog 程序...\n")
for i, statement in enumerate(prolog_model):
    print(f"语句 {i+1}:")
    
    try:
        if isinstance(statement, Clause):
            parse_clause(statement)
        else:
            parse_term(statement)
    except Exception as e:
        print(f"解析错误: {e}")
    
    print("\n" + "="*50 + "\n")

# 获取所有变量和常量（使用更安全的方法）
all_vars = set()
all_constants = set()

for statement in prolog_model:
    try:
        # 获取该语句中的所有变量
        if isinstance(statement, Clause):
            if hasattr(statement, 'variables'):
                for var in statement.variables():
                    all_vars.add(var)
            
            # 递归获取所有子项
            def collect_subterms(term):
                if is_variable(term):
                    all_vars.add(term)
                elif isinstance(term, Constant):
                    all_constants.add(term)
                elif isinstance(term, Term):
                    for arg in term.args:
                        collect_subterms(arg)
            
            # 处理头部
            if hasattr(statement, 'head'):
                collect_subterms(statement.head)
            
            # 处理体部
            if hasattr(statement, 'body') and statement.body is not None:
                collect_subterms(statement.body)
        
        # 处理简单项（如事实）
        elif isinstance(statement, Term):
            for arg in statement.args:
                collect_subterms(arg)
    except Exception as e:
        print(f"收集变量/常量时出错: {e}")

print("程序中的所有变量:")
for var in all_vars:
    print(f"  {var}")

print("\n程序中的所有常量:")
for const in all_constants:
    print(f"  {const}")
```

```python
from problog.program import PrologString
from problog.logic import Term, Clause, Constant, Var, is_variable
import re

def parse_program_with_comments(program_str):
    """
    解析ProbLog程序，同时提取注释并与代码关联
    """
    # 存储语句和相关注释
    statements_with_comments = []
    
    # 分割成行
    lines = program_str.split('\n')
    
    # 提取单行注释
    line_comments = {}
    for i, line in enumerate(lines):
        # 寻找行内注释
        comment_match = re.search(r'%(.*)$', line)
        if comment_match:
            # 存储注释内容（不包括%）
            comment = comment_match.group(1).strip()
            # 存储去除注释后的代码行
            code_line = line[:comment_match.start()].strip()
            line_comments[i] = {'code': code_line, 'comment': comment}
        else:
            # 没有注释的行
            code_line = line.strip()
            if code_line:  # 忽略空行
                line_comments[i] = {'code': code_line, 'comment': None}
    
    # 提取块注释 /* ... */
    block_comments = []
    in_block_comment = False
    block_start = -1
    block_content = ""
    
    for i, line in enumerate(lines):
        # 检查块注释开始
        if '/*' in line and not in_block_comment:
            in_block_comment = True
            block_start = i
            start_pos = line.find('/*')
            # 获取这一行注释开始前的代码
            pre_comment_code = line[:start_pos].strip()
            # 获取这一行注释内容
            block_content = line[start_pos+2:].strip()
            # 如果同一行内有注释结束
            if '*/' in block_content:
                end_pos = block_content.find('*/')
                block_content = block_content[:end_pos].strip()
                post_comment_code = line[start_pos+2+end_pos+2:].strip()
                in_block_comment = False
                block_comments.append({
                    'start_line': i,
                    'end_line': i,
                    'content': block_content,
                    'pre_code': pre_comment_code,
                    'post_code': post_comment_code
                })
                block_content = ""
        # 块注释中间的行
        elif in_block_comment and '*/' not in line:
            block_content += " " + line.strip()
        # 块注释结束的行
        elif in_block_comment and '*/' in line:
            end_pos = line.find('*/')
            block_content += " " + line[:end_pos].strip()
            post_comment_code = line[end_pos+2:].strip()
            in_block_comment = False
            block_comments.append({
                'start_line': block_start,
                'end_line': i,
                'content': block_content,
                'pre_code': "",  # 块注释前的代码已在开始行处理
                'post_code': post_comment_code
            })
            block_content = ""
    
    # 创建ProbLog程序对象
    prog = PrologString(program_str)
    
    # 创建语句和注释的映射
    for i, statement in enumerate(prog):
        # 尝试在原始代码中找到此语句
        statement_str = str(statement)
        statement_info = {
            'statement': statement,
            'statement_str': statement_str,
            'line_comments': [],
            'block_comments': []
        }
        
        # 查找相关的行注释
        for line_num, line_info in line_comments.items():
            # 简单匹配：检查语句是否在代码行中
            # 这种方法不完美，但可以作为起点
            if line_info['code'] and statement_str in line_info['code'] and line_info['comment']:
                statement_info['line_comments'].append({
                    'line': line_num,
                    'comment': line_info['comment']
                })
        
        # 查找相关的块注释
        for block_comment in block_comments:
            # 检查语句是否紧跟在块注释之后
            if block_comment['post_code'] and statement_str in block_comment['post_code']:
                statement_info['block_comments'].append({
                    'lines': (block_comment['start_line'], block_comment['end_line']),
                    'comment': block_comment['content']
                })
        
        statements_with_comments.append(statement_info)
    
    return statements_with_comments

def display_parsed_program(statements_with_comments):
    """
    显示带有注释的已解析程序
    """
    print("\n===== 程序解析结果（带注释）=====\n")
    
    for i, info in enumerate(statements_with_comments):
        print(f"语句 {i+1}: {info['statement_str']}")
        print(f"  类型: {type(info['statement']).__name__}")
        
        # 显示关联的行注释
        if info['line_comments']:
            print("  行注释:")
            for comment_info in info['line_comments']:
                print(f"    行 {comment_info['line']+1}: {comment_info['comment']}")
        
        # 显示关联的块注释
        if info['block_comments']:
            print("  块注释:")
            for comment_info in info['block_comments']:
                start, end = comment_info['lines']
                print(f"    行 {start+1}-{end+1}: {comment_info['comment']}")
        
        # 解析语句的详细结构
        if isinstance(info['statement'], Clause):
            print("  子句结构:")
            print(f"    头部: {info['statement'].head}")
            print(f"    体部: {info['statement'].body}")
        elif isinstance(info['statement'], Term):
            print("  项结构:")
            print(f"    函子: {info['statement'].functor}")
            print(f"    参数: {info['statement'].args}")
        
        print("-" * 50)


# 解析程序及其注释
parsed_program = parse_program_with_comments(model)

# 显示解析结果
display_parsed_program(parsed_program)

# 详细解析特定语句
print("\n===== 详细解析示例 =====\n")

for i, info in enumerate(parsed_program):
    # 选择一个有注释的子句进行详细解析
    if isinstance(info['statement'], Clause) and info['line_comments'] or info['block_comments']:
        print(f"详细解析语句: {info['statement_str']}")
        
        # 解析头部
        head = info['statement'].head
        print(f"头部: {head}")
        print(f"  函子: {head.functor}")
        print(f"  参数数量: {head.arity}")
        for j, arg in enumerate(head.args):
            print(f"  参数 {j+1}: {arg} (类型: {type(arg).__name__})")
        
        # 解析体部
        body = info['statement'].body
        if body:
            print(f"体部: {body}")
            if body.functor == 'conjunction':
                print(f"  这是一个AND连词，包含 {len(body.args)} 个子目标")
                for j, subgoal in enumerate(body.args):
                    print(f"  子目标 {j+1}: {subgoal}")
            else:
                print(f"  函子: {body.functor}")
                print(f"  参数数量: {body.arity}")
                for j, arg in enumerate(body.args):
                    print(f"  参数 {j+1}: {arg} (类型: {type(arg).__name__})")
        
        # 显示相关注释
        if info['line_comments']:
            print("相关行注释:")
            for comment in info['line_comments']:
                print(f"  {comment['comment']}")
        
        if info['block_comments']:
            print("相关块注释:")
            for comment in info['block_comments']:
                print(f"  {comment['comment']}")
        
        print("-" * 50)
        
        # 只详细解析一个语句作为示例
        break
```
